```python
import pandas as pd
```

```python
import numpy as np
```

```python
# reading dataset

RD = pd.read_csv("dataset.csv")
```

```python
RD
```

```python
# creating column Symptoms

RD["Symptoms"] = 0
```

```python
N = RD.shape[0]

for i in range(N):
    values = RD.iloc[i].values
    values = values.tolist()
    
    if 0 in values:
        RD["Symptoms"][i] = values[1:values.index(0)]
    else:
        RD["Symptoms"][i] = values[1:]
```

```python
RD.head()
```

```python
# finding all unique symptoms

col_values = RD[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4',
       'Symptom_5', 'Symptom_6', 'Symptom_7', 'Symptom_8', 'Symptom_9',
       'Symptom_10', 'Symptom_11', 'Symptom_12', 'Symptom_13', 'Symptom_14',
       'Symptom_15', 'Symptom_16', 'Symptom_17']].values.ravel()
```

```python
symp = pd.unique(col_values)

# creating list of symptoms

symp = symp.tolist()

# removing "nan" values

symp = [i for i in symp if str(i) != "nan"]
```

```python
# creating new dataset with unique symptoms as features

ProsDataset = pd.DataFrame(columns = symp, index = RD.index)

ProsDataset 
```

```python
#entering values in dataset

for i in symp:
    ProsDataset[i] = RD.apply(lambda x:1 if i in x.Symptoms else 0, axis = 1)
```

```python
#adding disease column to new dataset

ProsDataset["Disease"] = RD["Disease"]
```

```python
#proccessed data 

ProsDataset
```

```python
# saving proccessed dataset

ProsDataset.to_csv("Preproccessed dataset.csv")
```
